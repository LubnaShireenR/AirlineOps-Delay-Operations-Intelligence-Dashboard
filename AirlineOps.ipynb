```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("airline_dataset.csv")
```

```python
df.head()
df.shape
```

```python
df.columns.tolist()
df.info()
df.isnull().sum()
df.describe()
```

```python
df["Flight_Status"] = df["cancelled"].map({
    0: "Operated",
    1: "Cancelled"
})
```

```python
df["Flight_Status"].value_counts()
```

```python
df["Month_Name"] = pd.to_datetime(
    df["month"],
    format="%m"
).dt.month_name()
```

```python
df["Month_Name"].head()
```

```python
df["Weather_Delay_Category"] = pd.cut(
    df["weather_delay"],
    bins=[-1, 0, 30, 60, 1000],
    labels=[
        "No Delay",
        "Low Delay",
        "Medium Delay",
        "High Delay"
    ]
)
```

```python
df["Weather_Delay_Category"].value_counts()
```

```python
print("Total Flights:", len(df))

print("Cancelled Flights:",
      df["cancelled"].sum())

print("Average Distance:",
      round(df["distance"].mean(), 2))

print("Average Air Time:",
      round(df["air_time"].mean(), 2))
```

```python
df[[
    "air_time",
    "distance",
    "weather_delay",
    "late_aircraft_delay"
]].isnull().sum()
```

```python
df["weather_delay"] = df["weather_delay"].fillna(0)

df["late_aircraft_delay"] = df["late_aircraft_delay"].fillna(0)

df["air_time"] = df["air_time"].fillna(
    df["air_time"].median()
)
```

```python
df["Delay_Impact"] = (
    df["weather_delay"] +
    df["late_aircraft_delay"]
)
```

```python
df["Delay_Severity"] = pd.cut(
    df["Delay_Impact"],
    bins=[-1,0,30,60,10000],
    labels=[
        "No Delay",
        "Low",
        "Medium",
        "High"
    ]
)
```

```python
df["Delay_Severity"].value_counts()
```

```python
df["origin"].value_counts().head(10)
```

```python
monthly_flights = (
    df.groupby("Month_Name")
    .size()
    .reset_index(name="Flights")
)

monthly_flights
```

```python
print("Total Flights:", len(df))

print("Cancelled Flights:",
      df["cancelled"].sum())

print("Cancellation Rate:",
      round(df["cancelled"].mean()*100,2),
      "%")

print("Average Air Time:",
      round(df["air_time"].mean(),2))

print("Average Distance:",
      round(df["distance"].mean(),2))
```

```python
sample_df = df.sample(
    n=100000,
    random_state=42
)

sample_df.to_csv(
    "airlineops_tableau.csv",
    index=False
)

print("Tableau Dataset Created Successfully")
```
